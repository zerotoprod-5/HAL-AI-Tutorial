<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #1f2d3d;background:#eef2f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#1f2d3d;font-size:13px;margin-bottom:6px;text-transform:uppercase;">The scenario</div><b>Our scenario: a fleet of similar machines.</b><br>Picture a plant floor with hundreds of similar machines, each fitted with sensors. For our historical machines we also know the outcome: did it <b>fail soon</b> afterwards (within the next maintenance window), or did it keep running fine?<br><br>The question: <b>given a machine's recent sensor readings, can we predict whether it will fail soon &mdash; in time to service it on our terms instead of in an emergency?</b> No deep machinery knowledge required; the common-sense story is enough: hot, shaky, badly-pressured machines running on dirty oil that are overdue for service are the ones to worry about.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 1</span>The old way vs. predictive maintenance</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Predictive maintenance</div><b>Predictive maintenance</b> means using a machine's data to predict failure <i>before</i> it happens, so the fix is <b>planned</b>, not an emergency. <div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Remaining useful life</div>A closely related idea you will hear is <b>remaining useful life</b> &mdash; roughly, <i>how much life a machine has left before it fails</i>. Today we answer the simpler yes/no version ("will it fail soon?"), but it is the same family of question, and the same sensor data feeds both.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 2</span>Build the fleet's data</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Quick reminders</div>From Notebook 0: a <b>dataset</b> is a table (rows = machines, columns = measurements); the input columns are <b>features</b>; the answer column is the <b>label</b>. We will not re-teach these &mdash; just watch them in action on a richer, five-sensor table.</div>

```python
# Set up our toolboxes. This cell prints nothing - that is normal.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix

print('Tools loaded. We are ready to build the fleet.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>We loaded our toolboxes. Two new names to notice: <code>RandomForestClassifier</code> (the model we train first) and <code>IsolationForest</code> (the surprise at the end). More on both when we reach them.</div>

```python
# Create 600 historical machines. In real life this table is your sensor logs;
# here we generate it so every attendee has identical numbers.
rng = np.random.default_rng(42)   # fixed seed => everyone sees the same data
n = 600

# Five sensor readings per machine:
temperature        = np.round(rng.normal(70, 9, n), 1)                       # running temp (deg C)
vibration          = np.round(np.clip(rng.normal(3.0, 1.0, n), 0.2, None), 2) # mm/s
pressure           = np.round(rng.normal(5.0, 0.8, n), 2)                     # bar (normal is ~5)
oil_quality        = np.round(np.clip(rng.normal(75, 12, n), 5, 100), 0)      # 0-100, higher = cleaner
hours_since_service= rng.integers(50, 2000, n)                               # hours since last service

# The hidden truth we pretend not to know. Failure risk goes UP with:
#   high temperature, high vibration, abnormal pressure (far from 5 bar in EITHER
#   direction), LOW oil quality (dirty oil), and many hours since service.
risk = ( 0.060*(temperature - 70)
       + 0.95 *(vibration - 3.0)
       + 0.55 *np.abs(pressure - 5.0)
       + 0.045*(75 - oil_quality)
       + 0.0013*(hours_since_service - 1000)
       + rng.normal(0, 0.35, n) )        # real-world noise: the world is not tidy

will_fail_soon = (risk > 1.55).astype(int)   # 1 = failed soon after, 0 = kept running

fleet = pd.DataFrame({
    'temperature':         temperature,
    'vibration':           vibration,
    'pressure':            pressure,
    'oil_quality':         oil_quality,
    'hours_since_service': hours_since_service,
    'will_fail_soon':      will_fail_soon,
})

fleet.head(8)   # peek at the first 8 machines
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>There is our fleet: 600 machines, one per row. The first five columns are the <b>features</b> (the sensor readings); the last column, <code>will_fail_soon</code>, is the <b>label</b> &mdash; the outcome we want to predict. Notice <code>oil_quality</code> runs the other way: a <i>higher</i> number means <i>cleaner</i> oil, so <i>low</i> oil quality is the worrying one.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 3</span>How often do machines actually fail?</div>

Before modelling, always look at the balance of the label. If failures are rare, a lazy model could score well just by saying "everything is fine" &mdash; so we need to know the failure rate up front.

```python
print('Fleet size (machines, columns):', fleet.shape)
print()
counts = fleet['will_fail_soon'].value_counts().sort_index()
print('Outcome counts:')
print('  kept running (0):', counts.get(0, 0))
print('  failed soon  (1):', counts.get(1, 0))
print()
fail_rate = fleet['will_fail_soon'].mean()
print('Failure rate:', round(fail_rate*100, 1), '% of the fleet failed soon')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Only about <b>one machine in five</b> fails soon &mdash; failures are the <b>minority</b>. That single fact shapes everything that follows: a model that lazily declares every machine healthy would still be right ~80% of the time, yet would catch <b>zero</b> real failures. That is why, in a moment, accuracy alone will not be enough to trust the model.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 4</span>Split features from label, then hold back a test set</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #445566;background:#eef1f4;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#445566;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Note</div>Same rhythm as Notebook 0: separate the feature table <code>X</code> from the label column <code>y</code>, then hide a quarter of the machines as a <b>test set</b> so we can judge the model on cases it never saw. We keep <code>stratify=y</code> so the test set has the same ~20% failure rate as the whole fleet.</div>

```python
feature_cols = ['temperature', 'vibration', 'pressure', 'oil_quality', 'hours_since_service']
X = fleet[feature_cols]      # the clues  (five sensors)
y = fleet['will_fail_soon']  # the answer (failed soon or not)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,     # keep 25% hidden for the final test
    random_state=0,     # fixed split so everyone matches
    stratify=y,         # keep the same failure rate in both halves
)

print('Learn from :', X_train.shape[0], 'machines  (training set)')
print('Tested on  :', X_test.shape[0],  'machines  (test set, kept hidden)')
print('Failures in the hidden test set:', int(y_test.sum()))
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>450 machines go to training, 150 are locked away for the final test. Thanks to <code>stratify</code>, both halves carry the same ~20% failure rate, so the test is fair.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 5</span>Train the model: a random forest</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Random forest</div>In Notebook 0 we trained a single <b>decision tree</b> &mdash; one chain of yes/no questions. A <b>random forest</b> is simply <b>many small decision trees voting together</b>. Each tree sees a slightly different slice of the data and casts a vote; the forest goes with the majority. Many modest opinions combined beat one opinion &mdash; the same reason a review panel beats a lone reviewer. It is a workhorse of predictive maintenance.</div>

```python
# Build a forest of 200 trees and let it study the training machines.
model = RandomForestClassifier(
    n_estimators=200,           # 200 trees vote
    max_depth=8,                # let each tree learn a bit more detail
    class_weight='balanced',    # value rare failures more, so we catch them
    random_state=0,             # fixed so everyone gets the same forest
)

model.fit(X_train, y_train)   # <-- the learning happens here

print('Training done. 200 trees have studied the fleet and are ready to vote.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>That one <code>.fit()</code> call trained all 200 trees at once. The forest has now learned, on its own, how the five sensors combine into failure risk. One deliberate choice: <code>class_weight='balanced'</code> tells the forest to take the <i>rare</i> failures seriously rather than ignore them &mdash; the exact knob you turn when missing a fault is far costlier than a false alarm. Next we check &mdash; honestly &mdash; how good it is.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 6</span>Evaluate honestly: accuracy is not enough</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Quick reminders from Notebook 1</div><b>Recall</b> = of the machines that truly failed, what fraction did we catch? A <b>missed fault</b> is a real failure the model called healthy (dangerous &mdash; the machine breaks anyway). A <b>false alarm</b> is a healthy machine the model flagged (wasteful &mdash; we service it for nothing). The <b>confusion matrix</b> lays all four outcomes in one square. For rare, costly failures, <b>recall matters more than raw accuracy</b>.</div>

```python
predictions = model.predict(X_test)   # the forest's vote for each hidden machine

acc    = accuracy_score(y_test, predictions)
recall = recall_score(y_test, predictions)   # fraction of real failures we caught

print('Accuracy :', round(acc, 3),  '-> fraction of all machines called correctly')
print('Recall   :', round(recall, 3), '-> fraction of REAL failures we actually caught')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Accuracy is high (~0.9), but the number that should hold your attention is <b>recall</b>: it tells us how many of the genuinely failing machines we caught in time. On rare failures you can have great accuracy and still miss faults &mdash; so we always read recall alongside it.</div>

```python
# The confusion matrix: all four outcomes in one square.
cm = confusion_matrix(y_test, predictions)   # rows = truth, cols = prediction
tn, fp, fn, tp = cm.ravel()

print('Confusion matrix (rows = truth, columns = prediction):')
print('                     said FINE   said WILL FAIL')
print(f'  truly fine         {tn:>6}        {fp:>6}')
print(f'  truly will fail    {fn:>6}        {tp:>6}')
print()
print(f'Caught failures (good)        : {tp}')
print(f'MISSED faults  (dangerous)    : {fn}   <- these break anyway')
print(f'False alarms   (wasteful)     : {fp}   <- serviced for nothing')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Read it in plain words: of the real failures, the forest <b>caught most and missed only a few</b> &mdash; the missed ones are the dangerous corner (they fail anyway), and the false alarms are merely wasteful (we service a healthy machine). In real maintenance you would tune the model to push missed faults toward zero, even at the cost of a few more false alarms &mdash; a missed fault on the plant floor costs far more than an extra inspection.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 7</span>Which sensor mattered most?</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Feature importance</div>A model does not have to be a black box. <b>Feature importance</b> is a number, one per feature, telling you <b>how much the model leaned on each sensor</b> when making its calls. Bigger bar = the model found that sensor more useful. This is where predictive AI stops being a yes/no oracle and starts giving <b>engineers insight</b>: it points at <i>which signal</i> is driving failures across the fleet.</div>

```python
# Every trained forest exposes how much it relied on each feature.
importances = model.feature_importances_

ranking = (pd.Series(importances, index=feature_cols)
             .sort_values(ascending=True))   # ascending so the biggest bar lands on top

print('How much the model leaned on each sensor (higher = more):')
for name, val in ranking[::-1].items():
    print(f'  {name:<20} {val:.3f}')
```

```python
# Draw it as a horizontal bar chart - the ranking is easier to read this way.
plt.figure(figsize=(8, 5.5))
plt.barh(ranking.index, ranking.values, color='#0b6e7a', edgecolor='white')
plt.xlabel('Importance (share of the decision the model gave this sensor)')
plt.title('Which sensor drove the failure predictions?')
plt.grid(alpha=0.3, axis='x')
plt.tight_layout()
plt.show()
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>The forest tells us, in its own words, which sensors it trusted most &mdash; with <b>vibration</b> and <b>temperature</b> near the top, matching the common-sense story we started with. This is the payoff engineers love: the model is not just saying "this machine will fail," it is saying <b>which signal to watch</b>. If vibration dominates across your fleet, that is a direct hint about where to focus inspections &mdash; insight you can act on, discovered automatically from the data rather than assumed.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 8</span>Score three specific machines</div>

The whole point: a new machine reports its sensors, and we get an instant call plus a confidence. Let us run one clearly healthy machine, one clearly at-risk, and one genuinely borderline.

```python
# Three machines with very different sensor profiles.
new_machines = pd.DataFrame([
    # cool, smooth, normal pressure, clean oil, recently serviced -> should look healthy
    {'temperature': 64, 'vibration': 2.1, 'pressure': 5.0, 'oil_quality': 90, 'hours_since_service': 200},
    # hot, shaky, off-pressure, dirty oil, long overdue -> should look at-risk
    {'temperature': 88, 'vibration': 5.2, 'pressure': 6.6, 'oil_quality': 35, 'hours_since_service': 1900},
    # middling on everything -> genuinely borderline
    {'temperature': 73, 'vibration': 3.1, 'pressure': 5.2, 'oil_quality': 68, 'hours_since_service': 950},
], index=['Machine A (healthy?)', 'Machine B (at-risk?)', 'Machine C (borderline?)'])

guesses     = model.predict(new_machines)
confidences = model.predict_proba(new_machines)[:, 1]   # probability of 'will fail soon'

for name, guess, conf in zip(new_machines.index, guesses, confidences):
    verdict = 'WILL FAIL SOON' if guess == 1 else 'looks fine'
    print(f'{name:<26} -> {verdict:<16} (failure confidence {round(conf*100):>3} %)')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>The forest clears the cool/smooth machine, flags the hot/shaky/overdue one with high confidence, and lands the middling machine near the middle &mdash; exactly what an experienced engineer would say. The difference: the model does this for the whole fleet in a heartbeat, consistently, every shift. That confidence number is also how you <b>prioritise</b>: service the 92%-risk machines before the 55%-risk ones.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 9</span>What if we have NO failure labels?</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #445566;background:#eef1f4;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#445566;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Note</div>Everything above needed a <code>will_fail_soon</code> column &mdash; a record of what failed. But often you install sensors on a new fleet and have <b>only the readings, with no failure history at all</b>. You cannot train the model we just built, because there are no answers to learn from. So is AI useless here? Not at all &mdash; this is where the <i>other half</i> of machine learning begins.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Supervised vs unsupervised</div>Everything so far was <b>supervised</b> learning: we learned from labelled examples (each machine came with its known answer). <b>Unsupervised</b> learning has <b>no labels at all</b> &mdash; the model gets only the readings and must find structure on its own. No answer key, no teacher; just the data.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Anomaly detection</div>One powerful unsupervised tool is <b>anomaly detection</b>: find the machines that look <b>unusual compared to the normal crowd</b>, without ever being told what failure looks like. The idea is simple &mdash; most machines behave normally, so the few that sit far from the pack are the ones worth a closer look.</div>

```python
# Hand the model ONLY the sensors - we deliberately ignore will_fail_soon now,
# pretending we never had failure labels in the first place.
X_unlabelled = fleet[feature_cols]

detector = IsolationForest(
    contamination=0.08,   # assume ~8% of machines are unusual enough to flag
    random_state=0,
)
flag = detector.fit_predict(X_unlabelled)   # -1 = anomaly (unusual), 1 = normal

fleet_view = fleet.copy()
fleet_view['anomaly'] = (flag == -1)

n_flagged = int(fleet_view['anomaly'].sum())
print('Machines flagged as unusual:', n_flagged, 'out of', len(fleet_view))
print()
print('A few of the flagged (unusual) machines:')
print(fleet_view[fleet_view['anomaly']][feature_cols].head(6).to_string())
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>With <b>no labels whatsoever</b>, the model picked out the ~8% of machines that look least like the normal crowd. Glance at their numbers: they tend to sit at extremes &mdash; very hot, very shaky, very off-pressure. The model was never told what "bad" means; it simply found the <b>odd ones out</b>.</div>

```python
# See it: temperature vs vibration, normal machines green, flagged ones red X.
normal  = fleet_view[~fleet_view['anomaly']]
unusual = fleet_view[ fleet_view['anomaly']]

plt.figure(figsize=(8, 5.5))
plt.scatter(normal['temperature'], normal['vibration'],
            c='#2e7d32', alpha=0.6, edgecolors='white', label='Normal machine')
plt.scatter(unusual['temperature'], unusual['vibration'],
            c='#9c2b2b', marker='X', s=90, label='Flagged as unusual')

plt.xlabel('Temperature (deg C)')
plt.ylabel('Vibration (mm/s)')
plt.title('Anomaly detection found the odd machines on its own')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
