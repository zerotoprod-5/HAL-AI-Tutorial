A quick reminder of words you already met in Notebook 0: a <b>library</b> is a ready-made toolbox of code. <code>pandas</code> handles tables, <code>scikit-learn</code> is the machine-learning toolbox, <code>matplotlib</code> draws charts. We just borrow them again.

```python
# Import the toolboxes. Running this cell shows no output - that is normal.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

print('Tools loaded. We are ready.')
```

The one new face is <code>LinearRegression</code> &mdash; the model that predicts a <b>number</b> instead of a category. We meet it properly in Step 4.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;background:#1f2d3d;border-radius:10px;padding:28px 30px;color:#fff;"><div style="text-transform:uppercase;letter-spacing:2px;font-size:13px;color:#7fb6c4;font-weight:700;">Part A</div><div style="font-size:30px;font-weight:800;margin-top:6px;line-height:1.2;">Regression &mdash; predict a number</div><div style="color:#c7d2dd;font-size:15px;margin-top:8px;">Predict a machine's monthly running cost from its load level and its age</div></div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Regression</div><b>Regression</b> is predicting a <b>number on a sliding scale</b> rather than a category. <div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #1f2d3d;background:#eef2f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#1f2d3d;font-size:13px;margin-bottom:6px;text-transform:uppercase;">The scenario</div><b>Part A scenario: the monthly running-cost estimate.</b><br>A workshop runs many machines. Each one costs money every month &mdash; electricity, wear, upkeep. The finance team would love to <b>estimate that monthly cost up front</b> from two simple facts they always know about a machine: how hard it is worked (its <b>load level</b>, as a percentage) and how old it is (its <b>age in years</b>).<br><br>Common sense says a heavily-loaded, older machine costs more to run than a lightly-used new one. We will let the computer learn exactly <i>how much</i> more, from past cost records.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 2</span>Build the cost records</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Label (for regression)</div>Same word as Notebook 0, new flavour. The <b>label</b> is still the answer we want to predict &mdash; but here it is a <b>number</b> (<code>monthly_cost</code> in rupees), not a 1-or-0 category. The two input columns, <code>load_percent</code> and <code>age_years</code>, are still the <b>features</b> (the clues).</div>

```python
# We create 250 past machine cost records. In real life this would come from
# your accounts; here we generate it so everyone has identical data.
rng = np.random.default_rng(42)   # fixed seed => everyone sees the same numbers
n = 250

load_percent = rng.uniform(20, 100, n).round(1)   # how hard it is worked (%)
age_years    = rng.uniform(0, 15, n).round(1)      # how old it is (years)

# The hidden truth we pretend not to know: cost rises with both load and age,
# plus real-world noise (no two machines are ever identical).
monthly_cost = (3000 + 95*load_percent + 480*age_years
                + rng.normal(0, 1400, n)).round(0)   # rupees per month

df = pd.DataFrame({
    'load_percent': load_percent,
    'age_years':    age_years,
    'monthly_cost': monthly_cost,
})

df.head(8)   # show the first 8 machines
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>There is our <b>dataset</b>: 250 machines, one per row. The first two columns are the <b>features</b> (load and age, the clues). The last column, <code>monthly_cost</code>, is the <b>label</b> &mdash; the rupee number we want to predict. Notice it is a full number, not just 0 or 1.</div>

```python
# A quick look at typical values - the rupee costs span a wide range.
print('Rows and columns:', df.shape)
print()
print('Typical values per column:')
df.describe().round(0)
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Monthly cost runs from a few thousand rupees up to around twenty thousand, depending on the machine. That wide spread is exactly why a single average would be a poor guess &mdash; a young, lightly-loaded machine and an old, hard-worked one are nowhere near the same cost.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 3</span>See the trend with our own eyes</div>

Before any modelling, let us simply <i>draw</i> the data. We plot <b>age</b> against <b>cost</b>, one dot per machine. If our common-sense story holds, the dots should drift upward as we move right.

```python
plt.figure(figsize=(8, 5.5))
plt.scatter(df['age_years'], df['monthly_cost'],
            c='#0b6e7a', alpha=0.6, edgecolors='white')

plt.xlabel('Age of machine (years)')
plt.ylabel('Monthly running cost (Rs)')
plt.title('Each dot is one machine: older machines tend to cost more')
plt.grid(alpha=0.3)
plt.show()
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>The cloud of dots clearly <b>slopes upward</b> to the right: older machines tend to cost more to run. It is not a perfect straight march &mdash; the dots scatter, because load and plain randomness also matter &mdash; but the upward <b>trend</b> is unmistakable. That trend is the pattern a regression model will capture.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 4</span>Fit a line through the dots</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Line of best fit</div>Imagine laying a single straight ruler across that cloud of dots so it sits as <b>close as possible to all of them at once</b> &mdash; not touching every dot, but balanced through the middle of the crowd. That ruler is the <b>line of best fit</b>. The model's whole job in regression is to find the one best position for it. Once placed, the line gives a cost estimate for <i>any</i> age you point to.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Fit (training, for regression)</div>Same <code>.fit()</code> you already know. For regression, <b>fitting</b> means sliding and tilting that line until it sits as close as possible to the past dots. One line of code does it.</div>

```python
# For the picture, we first learn the line using ONE feature: age.
age_only = LinearRegression()
age_only.fit(df[['age_years']], df['monthly_cost'])   # find the best line

# Build the line's height across the full age range, to draw it.
age_grid  = pd.DataFrame({'age_years': np.linspace(0, 15, 100)})
line_cost = age_only.predict(age_grid)

plt.figure(figsize=(8, 5.5))
plt.scatter(df['age_years'], df['monthly_cost'],
            c='#0b6e7a', alpha=0.5, edgecolors='white', label='Past machines')
plt.plot(age_grid['age_years'], line_cost, c='#9c2b2b', linewidth=3, label='Line of best fit')

plt.xlabel('Age of machine (years)')
plt.ylabel('Monthly running cost (Rs)')
plt.title('The model finds the straight line closest to all the dots')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>There it is &mdash; the first big picture of this notebook. The red <b>line of best fit</b> runs straight through the middle of the dot cloud, rising with age. The computer was not told where to put it; it <b>slid the line into the position that sits closest to every dot at once</b>. Point at any age on the bottom axis, read the red line, and you have a cost estimate.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 5</span>Train the real model and measure its error</div>

The picture used age alone so we could see it on a flat chart. The real model uses <b>both</b> clues &mdash; load and age &mdash; together. We train it the usual way: split off a hidden test set, fit on the rest, then check it on machines it never saw.

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Error (MAE)</div>For categories we used <b>accuracy</b>. For numbers, accuracy makes no sense &mdash; we will almost never hit the cost to the exact rupee. Instead we ask: <b>on average, how far off are we?</b> The <b>Mean Absolute Error (MAE)</b> answers exactly that. An MAE of Rs 1,100 means: <i>typically, our estimate is off by about eleven hundred rupees</i> &mdash; high or low. Lower is better. It is in the same rupees as the thing we predict, so it reads in plain money.</div>

```python
# Use BOTH clues now. X = the features, y = the number we predict.
X = df[['load_percent', 'age_years']]   # the two clues
y = df['monthly_cost']                   # the number to predict

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0)   # hide 25% for an honest test

model = LinearRegression()
model.fit(X_train, y_train)   # <-- the model finds the best fit on past data

predictions = model.predict(X_test)   # estimate cost for the hidden machines
mae = mean_absolute_error(y_test, predictions)

print('Tested on', X_test.shape[0], 'machines the model had never seen.')
print('Average error (MAE): about Rs', round(mae))
print('In plain words: our cost estimate is typically off by ~Rs', round(mae), 'either way.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>On machines it had never seen, the model's cost estimate is typically off by only <b>around Rs 1,100</b> &mdash; against costs that run to twenty thousand. That is a genuinely useful estimate, and the small error is healthy: real costs have surprises, and a model that claimed zero error would be cheating somehow.</div>

```python
# A second way to SEE how good it is: plot predicted cost against the real cost.
# Perfect predictions would land exactly on the diagonal line.
plt.figure(figsize=(8, 5.5))
plt.scatter(y_test, predictions, c='#2e7d32', alpha=0.6, edgecolors='white',
            label='Test machines')

lo = min(y_test.min(), predictions.min())
hi = max(y_test.max(), predictions.max())
plt.plot([lo, hi], [lo, hi], c='#9c2b2b', linewidth=2, linestyle='--',
         label='Perfect prediction')

plt.xlabel('Actual monthly cost (Rs)')
plt.ylabel('Predicted monthly cost (Rs)')
plt.title('Points hug the diagonal => predictions are close to reality')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Each green dot is one test machine: its true cost across the bottom, our estimate up the side. The dots <b>cluster tightly along the dashed diagonal</b>, which is where a perfect estimate would land. Close to the line means a good estimate; the small scatter around it is exactly the ~Rs 1,100 average error in picture form.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 6</span>Estimate the cost of two brand-new machines</div>

This is the payoff. Two new machines arrive and we want their monthly cost <i>before</i> the bills come in. We hand each one's two numbers to the trained model and read back an instant rupee estimate.

```python
# A light, new machine: low load, almost brand new.
light_new = pd.DataFrame([{'load_percent': 30, 'age_years': 1}])
est1 = model.predict(light_new)[0]
print('Light + new machine :', light_new.to_dict('records')[0])
print('  Estimated monthly cost: about Rs', round(est1))
print()

# A heavy, old machine: high load, well-aged.
heavy_old = pd.DataFrame([{'load_percent': 95, 'age_years': 13}])
est2 = model.predict(heavy_old)[0]
print('Heavy + old machine :', heavy_old.to_dict('records')[0])
print('  Estimated monthly cost: about Rs', round(est2))
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>The model estimates the light, new machine at roughly <b>Rs 6,000</b> a month and the heavy, old one at around <b>Rs 18,000</b> &mdash; about three times as much. That ordering matches plain common sense, but now we have an actual rupee figure for each, produced in an instant and consistent across thousands of machines.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;background:#1f2d3d;border-radius:10px;padding:28px 30px;color:#fff;"><div style="text-transform:uppercase;letter-spacing:2px;font-size:13px;color:#7fb6c4;font-weight:700;">Part B</div><div style="font-size:30px;font-weight:800;margin-top:6px;line-height:1.2;">Forecasting &mdash; predict over time</div><div style="color:#c7d2dd;font-size:15px;margin-top:8px;">Forecast next year's spare-parts demand so the workshop can stock up in advance</div></div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #1f2d3d;background:#eef2f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#1f2d3d;font-size:13px;margin-bottom:6px;text-transform:uppercase;">The scenario</div><b>Part B scenario: planning the spare-parts shelf.</b><br>The workshop's store keeps spare parts on the shelf. Order too few and machines sit idle waiting; order too many and cash is tied up gathering dust. To plan well, the store manager wants to know: <b>roughly how many parts will we use each month over the coming year?</b><br><br>All they have is the past &mdash; three years of monthly usage records. We will read the shape of that history and <b>project it forward</b>.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Forecasting</div><b>Forecasting</b> is regression where the main clue is <b>time</b>. Instead of "given this machine's numbers, what is its cost?", we ask "given how demand has moved month after month, what comes <i>next</i>?" The answer is still a number; the input is the calendar.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Trend</div>A <b>trend</b> is the overall direction the numbers drift over time &mdash; slowly rising, slowly falling, or flat. If spare-parts usage creeps up a little every year as the fleet ages, that steady climb is the trend.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Seasonality</div><b>Seasonality</b> is a pattern that <b>repeats on a calendar</b> &mdash; the same bump at the same time, year after year. Maybe usage rises every monsoon, or before an annual overhaul. It is the regular wiggle that rides on top of the overall trend.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 7</span>Build and plot the demand history</div>

```python
# We create 36 months (3 years) of past spare-parts demand.
# In real life this is your store's issue register; here we generate it.
rng2 = np.random.default_rng(42)
months      = 36
month_index = np.arange(months)          # 0, 1, 2, ... 35
month_of_year = month_index % 12          # 0..11, the calendar position

# Hidden truth: a gentle upward TREND + a repeating SEASONAL wiggle + noise.
trend  = 120 + 4.2 * month_index
season = 22*np.sin(2*np.pi*month_of_year/12) + 12*np.cos(2*np.pi*month_of_year/12)
noise  = rng2.normal(0, 13, months)
demand = (trend + season + noise).round(0)   # parts used that month

hist = pd.DataFrame({'month_index': month_index, 'demand': demand})
print('First 6 months of history:')
print(hist.head(6).to_string(index=False))
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>We now have three years of monthly demand, with two ingredients baked in: a slow upward <b>trend</b> (the fleet uses a little more each year) and a repeating <b>seasonal</b> wiggle. The numbers below are what the store actually issued each month &mdash; let us plot them.</div>

```python
plt.figure(figsize=(8, 5.5))
plt.plot(hist['month_index'], hist['demand'],
         c='#0b6e7a', linewidth=2, marker='o', markersize=4)

plt.xlabel('Month number (0 = three years ago)')
plt.ylabel('Spare parts used')
plt.title('Three years of monthly spare-parts demand')
plt.grid(alpha=0.3)
plt.show()
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Two things jump out. The line <b>drifts upward</b> over the three years &mdash; that is the <b>trend</b>. And it <b>wiggles up and down in a repeating rhythm</b> rather than climbing smoothly &mdash; that is the <b>seasonality</b>, the same calendar bump returning each year. A good forecast has to capture both.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 8</span>Teach a model the trend and the season</div>

We use the same <code>LinearRegression</code> from Part A &mdash; no new tool. We give it three clues about each month: the <b>month number</b> (which carries the rising trend) and two <b>seasonal markers</b> that tell the model where we are in the yearly cycle. With those, one straight-line model can ride both the climb and the repeating wiggle.

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #445566;background:#eef1f4;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#445566;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Note</div>The two seasonal markers are just a standard way to hand the model a calendar position as numbers (a sine and cosine of the month). You do not need the maths &mdash; think of them as a dial that tells the model "we are <i>here</i> in the yearly cycle", so it can add the right seasonal bump.</div>

```python
# Turn a list of month numbers into the three clues the model reads.
def make_features(idx):
    moy = idx % 12                                   # position in the year (0..11)
    return np.column_stack([
        idx,                                          # month number  -> the trend
        np.sin(2*np.pi*moy/12),                       # seasonal marker 1
        np.cos(2*np.pi*moy/12),                       # seasonal marker 2
    ])

X_time = make_features(month_index)

ts_model = LinearRegression()
ts_model.fit(X_time, demand)   # learn trend + season from the 36 past months

fitted = ts_model.predict(X_time)
fit_mae = mean_absolute_error(demand, fitted)
print('The model has learned the history.')
print('On the past months it is off by only ~', round(fit_mae), 'parts on average.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>The model learned the shape of the past in one <code>.fit()</code>: it captured both the upward trend and the repeating seasonal bump, matching the historical months to within about <b>8 parts</b> on average. Now the interesting part &mdash; we ask it about months it has never seen: the future.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 9</span>Forecast the next 6 months</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Forecast horizon</div>The <b>forecast horizon</b> is how far ahead we predict. Here it is the next <b>6 months</b> (month numbers 36 to 41). We simply ask the trained model for those future months &mdash; it continues the trend and keeps applying the seasonal rhythm.</div>

```python
# The next 6 months carry on from where history stopped.
future_index = np.arange(months, months + 6)        # 36, 37, ... 41
future_X     = make_features(future_index)
forecast     = ts_model.predict(future_X).round(0)   # the projected demand

for m, f in zip(future_index, forecast):
    print('Month', m, '(future)  ->  forecast', int(f), 'parts')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>The model projects demand for each of the next six months, carrying the upward trend forward and still applying the seasonal wiggle. These are the numbers the store manager would stock against. One picture makes it land &mdash; let us draw history and forecast on the same chart.</div>

```python
plt.figure(figsize=(8, 5.5))

# History: solid teal line.
plt.plot(month_index, demand, c='#0b6e7a', linewidth=2, marker='o',
         markersize=4, label='History (what happened)')

# Forecast: dashed amber line, starting where history ends.
bridge_x = np.append(month_index[-1], future_index)
bridge_y = np.append(demand[-1],      forecast)
plt.plot(bridge_x, bridge_y, c='#b26a00', linewidth=2.5, linestyle='--',
         marker='s', markersize=5, label='Forecast (next 6 months)')

plt.axvline(month_index[-1], color='#9c2b2b', linewidth=1, alpha=0.5)
plt.text(month_index[-1]-0.3, demand.min(), 'today', color='#9c2b2b',
         ha='right', fontsize=11)

plt.xlabel('Month number')
plt.ylabel('Spare parts used')
plt.title('Forecast continues the trend past today')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>There is the second big picture of the notebook. The solid teal line is the real past; the dashed amber line is the model's <b>forecast</b>, picking up exactly where history stops and <b>carrying the same rising, wiggling pattern into the future</b>. That visual continuation is the whole idea of forecasting: project tomorrow from the shape of yesterday.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #445566;background:#eef1f4;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#445566;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Note</div>An honest caveat. A forecast is a <b>best guess, not a promise</b>. The further out we look, the less certain it gets &mdash; a sudden new contract or a supplier delay can bend the real line away from our dashed one. That is why real planners never order the forecast exactly; they add a <b>safety margin</b> on top, larger for months further ahead.</div>
