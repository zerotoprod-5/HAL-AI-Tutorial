<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 2</span>Make a demo spoken report (no microphone needed)</div>

To keep this notebook fully self-contained &mdash; no microphones, no uploading files &mdash; we will *manufacture* a spoken report from a written sentence. The computer reads our sentence aloud and saves it as an audio file. That file then stands in for a real technician's recording.

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Text-to-speech (TTS)</div><b>Text-to-speech</b> turns written words into spoken audio &mdash; the opposite of what we really care about today. We use it for one small reason only: to <b>fabricate a demo clip</b> so everyone has an identical recording to work with, with no microphone required. The real star of the notebook is the step <i>after</i> this.</div>

```python
from gtts import gTTS                 # text-to-speech tool (makes our demo clip)
from IPython.display import Audio     # lets us play sound inside the notebook

# This written sentence is what we will pretend a technician spoke out loud.
spoken_report = 'There is an oil leak near the hydraulic pump and the pressure is dropping.'

# Turn that sentence into speech and save it as an audio file named report.mp3
gTTS(spoken_report).save('report.mp3')

print('Created an audio file: report.mp3')
print('It is the spoken version of:')
print('   "' + spoken_report + '"')
```

```python
# Press play on the little audio bar below to HEAR our demo report.
Audio('report.mp3')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>We just created <code>report.mp3</code> and played it. That audio file is now our <b>stand-in for a real spoken report</b> &mdash; exactly the kind of clip a technician's handheld would produce. From the computer's point of view, it has no idea what words are inside; it only has sound. Our job next is to recover the words.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 3</span>Listen to the audio and turn it into text</div>

This is the genuinely new skill of the notebook: handing raw audio to the computer and getting back the words. We load a small listening model and ask it to transcribe our clip.

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Speech-to-text / ASR (automatic speech recognition)</div><b>Speech-to-text</b> &mdash; also called <b>ASR</b>, automatic speech recognition &mdash; is a model that <b>listens</b> to audio and writes down the words it hears. It is exactly what your phone does when you dictate a message. We use a model called <b>Whisper</b>, in its smallest <code>tiny</code> size so it runs quickly here.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #445566;background:#eef1f4;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#445566;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Note</div>The first time you run the next cell, Whisper downloads its <code>tiny</code> model (a few seconds). It then prints a small warning about running on CPU &mdash; that is harmless and expected on Colab. Just wait for the text.</div>

```python
import whisper

# Load the smallest Whisper model. 'tiny' is fast and fine for short, clear clips.
model = whisper.load_model('tiny')

# Ask it to listen to our audio file and write down the words.
result = model.transcribe('report.mp3')
transcript = result['text'].strip()    # the recovered text

print('What the computer HEARD (the transcript):')
print('   "' + transcript + '"')
print()
print('The original sentence we spoke:')
print('   "' + spoken_report + '"')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>The computer <b>listened to sound and produced text</b> &mdash; and it is (nearly) word-for-word the sentence we spoke. It was never told the words; it worked them out from the audio alone. That recovered <code>transcript</code> is now ordinary text, which means we can feed it straight into the text analytics from Notebook 4.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #9c2b2b;background:#f8ecec;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#9c2b2b;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Watch out</div>ASR is not perfect. With a clear voice and no background noise &mdash; like our demo &mdash; it is excellent. But strong accents, machine noise on the shop floor, mumbling, or unusual part names can make it <b>mis-hear</b> a word ("hydraulic" might come out as "high drawlic"). In the real world you check important transcripts, and a noisy clip will read less cleanly than this quiet studio-style one. Honest expectations beat magical ones.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 4</span>Teach the computer the three teams (text analytics)</div>

Now we rebuild the Notebook 4 idea, self-contained right here. We give the computer a small set of past spoken-style reports, each already labelled with the team that handled it. From those examples it learns which words point to which team. This is the same learn-from-past-examples pattern as every notebook today &mdash; only the data is words instead of numbers.

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Turning words into numbers (TF-IDF)</div>A model can only do maths on numbers, so first we convert each report into a row of numbers that records <b>which words it contains</b>. The standard tool for this is <code>TfidfVectorizer</code>. The idea in one line: words that are <b>rare and distinctive</b> (like "voltage" or "hydraulic") count for more than common filler words (like "the"). After this step, a sentence is just a row of numbers &mdash; and we already know how to predict from rows of numbers.</div>

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

# A small set of PAST spoken-style reports, each labelled with the team that owns it.
# In real life these would be thousands of historical reports from your logs.
reports = [
    # --- Electrical ---
    'the control panel keeps tripping the circuit breaker',
    'there is a burning smell from the wiring near the motor',
    'voltage reading is unstable and a fuse blew again',
    'the indicator lights are dead, looks like a power supply fault',
    'sparks coming from the electrical junction box',
    'the battery is not charging and the alternator seems faulty',
    'the circuit breaker tripped and the wiring is overheating',
    'an electrical short blew the fuse and the panel lost power',
    # --- Mechanical ---
    'loud grinding noise from the gearbox under load',
    'the bearing is worn out and the shaft is vibrating',
    'a belt has snapped and the fan is not turning',
    'metal fatigue cracks visible on the mounting bracket',
    'the rotor is misaligned and rubbing against the casing',
    'excessive vibration and a broken coupling on the drive',
    'the gearbox is grinding and the shaft bearing is worn',
    'a snapped belt and a seized bearing stopped the fan',
    # --- Hydraulic ---
    'oil leak near the hydraulic pump and pressure is dropping',
    'the hydraulic cylinder is slow and losing pressure',
    'fluid leaking from a cracked hydraulic hose',
    'the pump is not building enough hydraulic pressure',
    'a seal has failed and hydraulic fluid is everywhere',
    'low oil level in the hydraulic reservoir, valve is sticking',
    'the hydraulic hose is leaking oil and pressure keeps dropping',
    'oil is leaking from the pump seal and the cylinder lost pressure',
]
teams = (
    ['Electrical'] * 8 +   # first 8 reports are Electrical
    ['Mechanical'] * 8 +   # next 8 are Mechanical
    ['Hydraulic']  * 8     # last 8 are Hydraulic
)

print('We have', len(reports), 'past reports across 3 teams:')
print('   Electrical, Mechanical, Hydraulic  (8 examples each)')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>There is our labelled history: 24 short reports, each tagged with the team that handled it. Eight examples per team. Small, but enough to show the idea. The computer has not learned anything yet &mdash; we have only laid out the examples.</div>

```python
# Snap the two pieces together: words -> numbers (TF-IDF), then a classifier.
# make_pipeline just chains them so we can train and predict in one go.
classifier = make_pipeline(
    TfidfVectorizer(),
    LogisticRegression(max_iter=1000),
)

# Train: the model studies the 18 past reports and learns which words signal which team.
classifier.fit(reports, teams)

print('Training done. The computer has learned the words for each team.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>One <code>.fit()</code> call &mdash; the same word you saw in Notebook 0 &mdash; and the model has learned which words lean Electrical, which lean Mechanical, and which lean Hydraulic. It is ready to judge a brand-new report.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 5</span>Route the spoken report automatically</div>

Here is the payoff. We take the **transcript** from Step 3 &mdash; the text the computer recovered from the audio &mdash; and ask the trained classifier which team it belongs to. Speech became text; now text becomes a decision.

```python
# Feed the TRANSCRIBED text (from Step 3) into the trained classifier.
predicted_team = classifier.predict([transcript])[0]

# How confident is it? Show the probability for each team.
probs = classifier.predict_proba([transcript])[0]
labels = classifier.classes_

print('Spoken report (transcribed):')
print('   "' + transcript + '"')
print()
print('Routed to team: ', predicted_team)
print()
print('Confidence per team:')
for team, p in sorted(zip(labels, probs), key=lambda kv: -kv[1]):
    bar = '#' * int(round(p * 30))    # a little text bar chart
    print('   {:<11} {:>4.0f}%  {}'.format(team, p * 100, bar))
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>The full pipeline just ran end to end: a <b>spoken</b> report became <b>text</b>, and that text was automatically <b>routed to the Hydraulic team</b> &mdash; with a confidence score to back it. Nobody typed, nobody sorted. The mention of an oil leak and hydraulic pressure made the answer clear, and the computer saw exactly those words.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 6</span>Try a different report, to prove it is not a fluke</div>

One example could be luck. Let us speak a *different* report &mdash; an electrical one this time &mdash; and run the whole pipeline again: make audio, transcribe, route. We bundle the three steps into one small helper so it is easy to reuse.

```python
def route_spoken_report(sentence):
    '''Speak the sentence, transcribe it, and route it to a team.'''
    # Step A: turn the sentence into audio (our stand-in for a recording).
    gTTS(sentence).save('clip.mp3')
    # Step B: transcribe the audio back into text.
    heard = model.transcribe('clip.mp3')['text'].strip()
    # Step C: route the transcribed text to a team.
    team = classifier.predict([heard])[0]
    confidence = classifier.predict_proba([heard]).max()
    print('Spoke      :', sentence)
    print('Transcribed:', heard)
    print('Routed to  :', team, '  ({:.0f}% confident)'.format(confidence * 100))
    return team

# A clearly electrical complaint this time.
route_spoken_report('The circuit breaker keeps tripping and there is a burning smell from the wiring.')
```

The words "circuit breaker", "burning smell", and "wiring" pointed there clearly. The pipeline <code>speak &rarr; transcribe &rarr; route</code> is reusable for any report &mdash; that one helper does it all.</div>
