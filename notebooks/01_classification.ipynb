<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #1f2d3d;background:#eef2f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#1f2d3d;font-size:13px;margin-bottom:6px;text-transform:uppercase;">The scenario</div><b>Our scenario: the final quality-control gate.</b><br>A workshop finishes a batch of parts. Before they ship, every part passes through a final <b>quality-control check</b>. Each part has a few simple measurements taken: how far its size is off target (<b>dimension error</b>), how rough its <b>surface</b> is, and how far its <b>weight</b> is from spec. Most parts are fine and should <b>PASS</b>. A minority are out of spec and must be <b>REJECTED</b> before they reach a customer.<br><br>The question: <b>can the computer look at a part's measurements and predict PASS or REJECT</b>, so good parts flow through and bad ones get caught? No special domain knowledge needed &mdash; bigger errors, rougher surfaces and larger weight deviations mean a part is more likely out of spec. That common-sense pattern is all we lean on.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 1</span>Re-introduce the task by name</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Classification</div><b>Classification</b> is predicting which <b>category</b> a case belongs to. The answer is not a number on a scale &mdash; it is a choice from a short list of labels. Here the list has two entries: <b>PASS</b> or <b>REJECT</b>. "Spam or not spam", "approve or decline", "defective or fine" are all classification.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Classes</div>The possible answers are called the <b>classes</b>. Our two classes are <b>PASS</b> (we will store it as <code>0</code>) and <b>REJECT</b> (stored as <code>1</code>). Computers prefer numbers to words, so we label the category we care about catching &mdash; REJECT &mdash; as <code>1</code>.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 2</span>Set up our tools</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Library</div>A quick reminder from Notebook 0: a <b>library</b> is a ready-made toolbox of code. <code>pandas</code> handles tables, <code>scikit-learn</code> is the machine-learning toolbox, <code>matplotlib</code> draws charts. We just borrow them.</div>

```python
# Import the toolboxes we need. This cell shows no visible output - that is
# normal. It only makes the tools available to the cells below.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, confusion_matrix,
                             precision_score, recall_score)

print('Tools loaded. We are ready.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>We loaded the toolboxes. Nothing dramatic on screen, and that is expected &mdash; this cell only prepares the tools the rest of the notebook uses.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 3</span>Build the data: 400 past parts</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Dataset and features</div>Reminder from Notebook 0: a <b>dataset</b> is just a table &mdash; one row per part. The input columns (<code>dimension_error_mm</code>, <code>surface_roughness</code>, <code>weight_deviation_g</code>) are the <b>features</b>, the clues the model gets to look at. The answer column is the <b>label</b>.</div>

```python
# We create 400 past parts that already went through quality control, so we
# know each one's real outcome. In real life this table would come from your
# inspection logs; here we generate it so everyone sees identical numbers.
rng = np.random.default_rng(42)   # fixed seed => same data for everyone
n = 400

# Three simple measurements per finished part:
dimension_error_mm = np.round(np.abs(rng.normal(0.0, 0.05, n)), 3)            # size off target (mm)
surface_roughness  = np.round(np.clip(rng.normal(1.2, 0.4, n), 0.2, None), 2) # roughness (Ra microns)
weight_deviation_g = np.round(rng.normal(0.0, 4.0, n), 1)                     # weight off target (grams)

# The hidden truth we pretend not to know: bigger size errors, rougher
# surfaces and larger weight deviations push a part toward being out of spec
# (plus a little real-world noise, because inspection is never perfectly clean).
risk = (35.0*dimension_error_mm + 2.4*(surface_roughness - 1.2)
        + 0.14*np.abs(weight_deviation_g) + rng.normal(0, 0.32, n))
status = (risk > 2.95).astype(int)   # 1 = REJECT (out of spec), 0 = PASS

df = pd.DataFrame({
    'dimension_error_mm': dimension_error_mm,
    'surface_roughness':  surface_roughness,
    'weight_deviation_g': weight_deviation_g,
    'status':             status,
})

df.head(8)   # show the first 8 parts
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>There is our <b>dataset</b>: 400 parts, one per row. The first three columns are the <b>features</b> (the measurements). The last column, <code>status</code>, is the <b>label</b> &mdash; <code>0</code> for PASS, <code>1</code> for REJECT &mdash; the outcome we want to predict for a new part.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 4</span>Look at the class balance</div>

Before any modelling, always look at the data. For classification the most important first question is: *how many of each class do we have?* The answer changes everything about how we judge the model later.

```python
print('Rows and columns:', df.shape)
print()

counts = df['status'].value_counts().sort_index()
print('How many PASS vs REJECT:')
print('  PASS   (0):', counts[0])
print('  REJECT (1):', counts[1])
print()

reject_share = 100 * df['status'].mean()
print('REJECT is the minority class:', round(reject_share, 1), '% of all parts.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #9c2b2b;background:#f8ecec;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#9c2b2b;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Watch out</div>Only about <b>20%</b> of parts are REJECT &mdash; the class is <b>imbalanced</b>. This is completely realistic (most finished parts are fine) but it is also a trap waiting to spring. Hold this number in your head: <b>roughly 4 out of every 5 parts PASS</b>. We will see in Step 7 exactly why that matters.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 5</span>See the two classes with our own eyes</div>

Before any maths, let us simply *draw* the parts. We plot two of the measurements against each other and colour each part green for PASS and red for REJECT. If our common-sense story holds, the colours should sit in different regions.

```python
plt.figure(figsize=(8, 5.5))

good = df[df['status'] == 0]   # the parts that PASSED
bad  = df[df['status'] == 1]   # the parts that were REJECTED

plt.scatter(good['dimension_error_mm'], good['surface_roughness'],
            c='#2e7d32', label='PASS',   alpha=0.7, edgecolors='white')
plt.scatter(bad['dimension_error_mm'],  bad['surface_roughness'],
            c='#9c2b2b', label='REJECT', alpha=0.7, edgecolors='white')

plt.xlabel('Dimension error (mm) - how far off target size')
plt.ylabel('Surface roughness (Ra microns)')
plt.title('Each dot is one finished part')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Notice two things. First, the red REJECT dots cluster toward the <b>top-right</b> &mdash; larger size error and rougher surface &mdash; exactly the common-sense pattern. Finding a boundary between the colours is the model's whole job. Second, look how <b>few</b> red dots there are compared to green. That scarcity is the imbalance from Step 4, and it is about to matter a great deal.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 6</span>Split features from label, then hold back a test set</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; X / y and train / test split</div>Reminders from Notebook 0: we call the feature table <code>X</code> (the clues) and the label column <code>y</code> (the answers). We then split the parts into a <b>training set</b> the model learns from, and a hidden <b>test set</b> used only at the end to score it on parts it has never seen. We use <code>stratify=y</code> so the rare REJECT class appears in the same proportion in both halves.</div>

```python
X = df[['dimension_error_mm', 'surface_roughness', 'weight_deviation_g']]   # the clues
y = df['status']                                                           # the answers

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,    # keep 25% (100 parts) hidden for the final test
    random_state=0,    # fixed split so everyone gets the same result
    stratify=y,        # keep the same PASS/REJECT mix in both halves
)

print('Learn from :', X_train.shape[0], 'parts  (training set)')
print('Tested on  :', X_test.shape[0],  'parts  (test set, kept hidden)')
print('REJECT parts hidden in the test set:', int(y_test.sum()),
      'out of', len(y_test))
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>300 parts go into the training set and 100 are locked away as the test set. Of those 100 hidden parts, only about 20 are genuine REJECTs &mdash; the rest PASS. Remember that count: it is the heart of the next step.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 7</span>Train the classifier</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Model and training (fitting)</div>A <b>model</b> is the pattern-finder; <b>training</b> &mdash; the method literally called <code>.fit()</code> &mdash; is where it studies the training parts and discovers the pattern. We use <b>logistic regression</b>, a classic, fast, well-understood classifier. (The <code>StandardScaler</code> in front just puts the three measurements on a comparable scale first, since they are in very different units &mdash; thousandths of a mm vs grams.)</div>

```python
# A pipeline = scale the measurements, then fit the classifier, as one unit.
model = make_pipeline(StandardScaler(),
                      LogisticRegression(max_iter=1000))

model.fit(X_train, y_train)   # <-- the model studies the training parts

print('Training done. The model has found its PASS / REJECT pattern.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>That one <code>.fit()</code> call is the entire act of learning. The model has now worked out how the three measurements combine into a PASS-or-REJECT decision. Next we put it to the test &mdash; and meet the most important lesson in this notebook.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 8</span>The accuracy trap: a lazy model that looks great</div>

Here is the lesson that separates people who really understand classification from those who just quote a score. Before we even look at our trained model, let us build a deliberately <b>lazy</b> model &mdash; one that does no thinking at all and simply predicts <b>PASS for every single part</b>. Then we measure its accuracy.

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Accuracy</div><b>Accuracy</b> is the simplest score: out of all test parts, what fraction did the model label correctly? <code>0.80</code> means 80 out of 100 right. It sounds like the obvious way to judge a model &mdash; and for imbalanced problems it can be dangerously misleading.</div>

```python
# The lazy model: ignore the measurements, just say PASS (0) for everything.
lazy_predictions = np.zeros(len(y_test), dtype=int)   # all zeros = all PASS

lazy_accuracy = accuracy_score(y_test, lazy_predictions)
print('Lazy model accuracy (always predicts PASS):', round(lazy_accuracy, 3))
print('In plain words: about', round(lazy_accuracy*100),
      'out of every 100 labelled correctly.')
print()

# But how many of the genuinely bad parts did this lazy model catch?
rejects_caught = int(((lazy_predictions == 1) & (y_test == 1)).sum())
rejects_total  = int((y_test == 1).sum())
print('REJECT parts it actually caught:', rejects_caught, 'out of', rejects_total)
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #9c2b2b;background:#f8ecec;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#9c2b2b;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Watch out</div>Read that result slowly. A model that does <b>no thinking whatsoever</b> &mdash; it just stamps PASS on everything &mdash; scores about <b>80% accuracy</b>. It looks like a B-grade student. But it caught <b>ZERO</b> of the bad parts. Every single out-of-spec part sailed straight through to the customer. On an imbalanced problem, <b>high accuracy can be worthless</b>: you can ace it by ignoring the very thing you were built to find. Accuracy alone is not enough &mdash; we need a tool that shows us <i>which kinds</i> of mistakes are happening.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 9</span>The confusion matrix: where mistakes actually happen</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Confusion matrix</div>A <b>confusion matrix</b> is a small table that splits the predictions into four buckets, so we can see not just <i>how many</i> the model got wrong but <i>which kind</i> of wrong. For a PASS/REJECT problem it is a 2&times;2 grid: real answer down the side, the model's call across the top.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; False alarm / Missed fault</div>Two of the four buckets are the mistakes, and they are <b>not</b> equally serious:<br><b>False alarm</b> &mdash; a <i>good</i> part the model wrongly REJECTED. Annoying and wasteful (you scrap a fine part), but safe. (Statisticians call this a <i>false positive</i>.)<br><b>Missed fault</b> &mdash; a <i>bad</i> part the model wrongly PASSED. This is the dangerous one: a defective part reaches the customer. (A <i>false negative</i>.)</div>

```python
# Now use our REAL trained model on the hidden test parts.
predictions = model.predict(X_test)

# confusion_matrix returns the four buckets. We unpack them by name so the
# meaning is crystal clear (tn, fp, fn, tp).
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()

print('Out of', len(y_test), 'hidden parts, the model produced:')
print('  True passes  (good part, correctly PASSED) :', tn)
print('  True rejects (bad part,  correctly REJECTED):', tp)
print('  False alarms (good part, wrongly REJECTED) :', fp)
print('  Missed faults(bad part,  wrongly PASSED)   :', fn)
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Unlike the lazy model, our trained classifier <b>actually catches most of the bad parts</b> &mdash; that is the <i>true rejects</i> number. It is not flawless: it raised a small number of <i>false alarms</i> (good parts it scrapped needlessly) and, more worryingly, let a few <i>missed faults</i> slip through. Now let us draw the same four numbers as a labelled grid, which is how engineers usually read them.</div>

```python
# Draw the confusion matrix as a labelled 2x2 grid.
cm = confusion_matrix(y_test, predictions)

fig, ax = plt.subplots(figsize=(8, 5.5))
ax.imshow(cm, cmap='Blues')

# Friendly labels instead of bare 0s and 1s.
ax.set_xticks([0, 1]); ax.set_xticklabels(['Model said\nPASS', 'Model said\nREJECT'])
ax.set_yticks([0, 1]); ax.set_yticklabels(['Really\nPASS', 'Really\nREJECT'])
ax.set_xlabel('What the model predicted')
ax.set_ylabel('The real answer')
ax.set_title('Confusion matrix: the four buckets')

# Write the count and a plain-English name inside each of the four cells.
names = [['True passes', 'False alarms'],
         ['Missed faults', 'True rejects']]
for i in range(2):
    for j in range(2):
        shade = 'white' if cm[i, j] > cm.max()/2 else '#1f2d3d'
        ax.text(j, i, f'{cm[i, j]}\n{names[i][j]}',
                ha='center', va='center', color=shade, fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>This single grid tells the whole story. The two cells on the <b>diagonal</b> (top-left and bottom-right) are the correct calls &mdash; true passes and true rejects. The two <b>off-diagonal</b> cells are the mistakes: top-right is the false alarms (good parts wrongly rejected), bottom-left is the missed faults (bad parts wrongly passed). For a safety gate, that bottom-left number is the one that should worry you most.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 10</span>Precision and recall: two honest scores</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Precision and recall</div>Two numbers that say far more than accuracy, in plain words &mdash; no formula to memorise:<br><b>Precision</b> answers: <i>"when the model shouts REJECT, how often is it actually right?"</i> Low precision means lots of false alarms &mdash; you scrap good parts.<br><b>Recall</b> answers: <i>"of all the genuinely bad parts, how many did the model actually catch?"</i> Low recall means lots of missed faults &mdash; bad parts escape.</div>

```python
precision = precision_score(y_test, predictions)   # of the REJECT calls, how many were right
recall    = recall_score(y_test, predictions)      # of the truly bad parts, how many we caught
accuracy  = accuracy_score(y_test, predictions)

print('Our trained model on the hidden test parts:')
print('  Accuracy :', round(accuracy, 3), ' (overall fraction labelled correctly)')
print('  Precision:', round(precision, 3),
      ' -> when it says REJECT, it is right', round(precision*100), '% of the time')
print('  Recall   :', round(recall, 3),
      ' -> it caught', round(recall*100), '% of all the truly bad parts')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #445566;background:#eef1f4;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#445566;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Note</div>Precision and recall usually pull against each other &mdash; this is the central <b>trade-off</b> in classification. If you make the model quicker to shout REJECT, it catches more bad parts (recall up) but also scraps more good ones (precision down), and vice-versa. <b>Which one matters more depends entirely on the cost of each mistake.</b> For a final safety gate, a missed fault (a defective part reaching a customer) is far more costly than a false alarm (scrapping a good part), so here we would care most about <b>recall</b> &mdash; catching the bad parts &mdash; even if it costs us a few extra false alarms.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Notice that accuracy and recall tell different stories. The lazy model and our real model have fairly similar <i>accuracy</i>, but their <i>recall</i> is worlds apart &mdash; the lazy model's recall is zero, ours catches most bad parts. That gap is exactly why engineers report precision and recall, not just accuracy, on any problem where the thing they must catch is rare.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 11</span>Use it on brand-new parts</div>

This is the payoff. Three new parts come off the line. We have their three measurements but no idea of the verdict. We hand each to the trained model and get an instant call, plus a <b>confidence</b> &mdash; the model's own probability that the part should be rejected.

```python
# Three new parts: one clearly good, one clearly bad, one genuinely borderline.
new_parts = pd.DataFrame([
    {'dimension_error_mm': 0.010, 'surface_roughness': 0.70, 'weight_deviation_g': 1.5},  # clearly good
    {'dimension_error_mm': 0.090, 'surface_roughness': 2.10, 'weight_deviation_g': 9.0},  # clearly bad
    {'dimension_error_mm': 0.050, 'surface_roughness': 1.55, 'weight_deviation_g': 5.0},  # borderline
])

calls       = model.predict(new_parts)              # 0 = PASS, 1 = REJECT
confidences = model.predict_proba(new_parts)[:, 1]  # probability of REJECT

labels = ['clearly good', 'clearly bad', 'borderline']
for i in range(len(new_parts)):
    verdict = 'REJECT' if calls[i] == 1 else 'PASS'
    print(f'{labels[i]:13s} -> {verdict:6s}   '
          f'(confidence it is a REJECT: {round(confidences[i]*100)}%)')
```

The confidence number is useful in practice: you might auto-PASS very low scores, auto-REJECT very high ones, and send only the uncertain middle to a human inspector.</div>
