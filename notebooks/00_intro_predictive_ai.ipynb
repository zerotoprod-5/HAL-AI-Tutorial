<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #1f2d3d;background:#eef2f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#1f2d3d;font-size:13px;margin-bottom:6px;text-transform:uppercase;">The scenario</div><b>Our running example: equipment servicing.</b><br>Imagine a workshop that maintains a fleet of identical machines. Over the years they have kept simple records for each machine: how many <b>hours</b> it has run, its running <b>temperature</b>, and how much it <b>vibrates</b>. For older records they also know the outcome: did that machine eventually <b>need a service</b>, or not?<br><br>The question we want answered: <b>given a machine's numbers, can we predict whether it will need servicing &mdash; before it breaks down?</b> No aircraft or engine knowledge required. Hotter, older, shakier machines tend to need attention. That common-sense pattern is all we rely on.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 1</span>Set up our tools</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Library</div>A <b>library</b> is a ready-made toolbox of code someone else wrote, so we do not start from scratch. <code>pandas</code> handles tables, <code>scikit-learn</code> is the machine-learning toolbox, <code>matplotlib</code> draws charts. We just borrow them.</div>

```python
# Import the toolboxes we need. Running this cell produces no visible output -
# that is normal. It just makes the tools available below.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import accuracy_score

print('Tools loaded. We are ready.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>We loaded our toolboxes. Nothing dramatic happened on screen, and that is expected &mdash; this cell only prepares the tools we use further down.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 2</span>Get the data</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Dataset</div>A <b>dataset</b> is just a <b>table</b>. Rows are the things we recorded (here, machines). Columns are the things we measured about each one. Nothing more exotic than a spreadsheet.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Feature</div>Each input column &mdash; <code>hours_used</code>, <code>temperature</code>, <code>vibration</code> &mdash; is called a <b>feature</b>. Features are the clues the computer gets to look at. One row of features describes one machine.</div>

```python
# We create 300 past machine records. In real life this table would come from
# your maintenance logs; here we generate it so everyone has identical data.
rng = np.random.default_rng(42)   # fixed seed => everyone sees the same numbers
n = 300

hours_used  = rng.integers(200, 6000, n)                 # running hours
temperature = np.round(rng.normal(72, 10, n), 1)          # running temp (deg C)
vibration   = np.round(np.clip(rng.normal(3.0, 1.0, n), 0.3, None), 2)  # mm/s

# The hidden truth we are pretending not to know: hotter + shakier + older
# machines are more likely to need a service (plus a little real-world noise).
risk = (0.00022*hours_used + 0.07*(temperature-72) + 0.85*(vibration-3.0)
        + rng.normal(0, 0.45, n))
needs_service = (risk > 0.55).astype(int)   # 1 = needed service, 0 = did not

df = pd.DataFrame({
    'hours_used':    hours_used,
    'temperature':   temperature,
    'vibration':     vibration,
    'needs_service': needs_service,
})

df.head(8)   # show the first 8 rows
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>There is our <b>dataset</b>: 300 machines, one per row. The first three columns are the <b>features</b> (the clues). The last column, <code>needs_service</code>, is the outcome we eventually want to predict.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 3</span>Look before you model</div>

Good practice: always *look* at the data before doing anything clever with it. How big is it? Is it balanced? What are typical values?

```python
print('Rows and columns:', df.shape)        # (machines, columns)
print()
print('How many needed service vs not:')
print(df['needs_service'].value_counts())
print()
print('Typical values per column:')
df.describe().round(1)
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Label</div>The outcome we want to predict &mdash; <code>needs_service</code>, which is 1 or 0 &mdash; is called the <b>label</b>. It is the "answer" for each past machine. Because we already know the answer for these old records, the computer can learn from them.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Label column</div>The single column that holds those answers is the <b>label column</b>. Separating the <b>label column</b> from the <b>feature columns</b> is the first real step of almost every predictive-AI project. Keep this distinction &mdash; it comes back in every notebook today.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>About half the machines needed service and half did not, so the data is nicely <b>balanced</b> &mdash; the computer will see plenty of both kinds. We also now know the typical hours, temperature and vibration.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 4</span>See the pattern with our own eyes</div>

Before any maths, let us simply *draw* the data. We will plot temperature against vibration, and colour each machine by whether it needed service. If our common-sense story is right, the two colours should fall in different regions.

```python
plt.figure(figsize=(8, 5.5))
ok   = df[df['needs_service'] == 0]
fail = df[df['needs_service'] == 1]

plt.scatter(ok['temperature'],   ok['vibration'],   c='#2e7d32', label='Did NOT need service', alpha=0.7, edgecolors='white')
plt.scatter(fail['temperature'], fail['vibration'], c='#9c2b2b', label='Needed service',        alpha=0.7, edgecolors='white')

plt.xlabel('Temperature (deg C)')
plt.ylabel('Vibration (mm/s)')
plt.title('Each dot is one machine')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Look at the colours: machines that needed service (red) cluster toward the <b>top-right</b> &mdash; hotter and shakier. Machines that were fine (green) sit toward the bottom-left. <b>That separation IS the pattern.</b> The whole job of a predictive model is to find a boundary between the colours, so that when a new grey dot arrives, it can guess the colour.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 5</span>Split the features from the label (X and y)</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; X and y</div>By long-standing convention we call the table of <b>features</b> <code>X</code> (the clues) and the <b>label</b> column <code>y</code> (the answers). The model's job is to learn the link <b>X &rarr; y</b>.</div>

```python
X = df[['hours_used', 'temperature', 'vibration']]   # the clues  (features)
y = df['needs_service']                               # the answers (label)

print('X is the feature table:', X.shape, '-> 300 machines, 3 clues each')
print('y is the label column: ', y.shape, '-> 300 answers')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>We split the one table into two parts: <code>X</code> holds the three clue columns, <code>y</code> holds the single answer column. Everything from here on works with these two.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 6</span>Hold back a test set</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Training set and test set</div>We split our 300 machines into two groups. The <b>training set</b> is what the model learns from. The <b>test set</b> is hidden away and used only at the end, to check the model on machines it has never seen.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #445566;background:#eef1f4;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#445566;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Note</div>Why hide some data? Think of a student. If you let them study the exact exam paper beforehand, a top score proves nothing. The honest test uses <b>questions they have not seen</b>. Same here: we judge the model only on the held-back test set.</div>

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,    # keep 25% hidden for the final test
    random_state=0,    # fixed split so everyone gets the same result
)

print('Learn from   :', X_train.shape[0], 'machines  (training set)')
print('Tested on    :', X_test.shape[0],  'machines  (test set, kept hidden)')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>225 machines go into the training set for the model to learn from, and 75 are locked away as the test set. The model will not see those 75 until the very end.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 7</span>Train the model</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Model</div>A <b>model</b> is the pattern-finder. Before training it knows nothing. We will use a <b>decision tree</b> &mdash; it learns a set of yes/no questions ("is vibration above 3?") that sort machines into "needs service" or "fine". It mirrors how a technician actually reasons.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Training (fitting)</div><b>Training</b> &mdash; the method is literally called <code>.fit()</code> &mdash; is the moment the model studies the training machines and discovers the pattern. This single line is where the "learning" in machine learning happens.</div>

```python
model = DecisionTreeClassifier(max_depth=3, random_state=0)

model.fit(X_train, y_train)   # <-- the model studies the training machines

print('Training done. The model has found its pattern.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>That one <code>.fit()</code> call is the entire act of learning. The model has now built its internal set of questions. Let us actually read them out.</div>

```python
# A decision tree is rare in that we can print the exact rules it invented.
print(export_text(model, feature_names=['hours_used', 'temperature', 'vibration']))
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>The computer <b>wrote its own if-then rules</b> from the data &mdash; nobody programmed them by hand. Read the top split: it first checks <code>vibration</code>, then <code>temperature</code>, then <code>hours_used</code> &mdash; exactly the common-sense clues we expected, discovered automatically.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 8</span>Predict, and measure how good it is</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Prediction</div>A <b>prediction</b> is the model's answer for a case: 1 (will need service) or 0 (fine). We ask it to predict the 75 hidden test machines, where we secretly know the real answers.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #0b6e7a;background:#e6f4f6;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#0b6e7a;font-size:13px;margin-bottom:6px;text-transform:uppercase;">Vocabulary &nbsp;—&nbsp; Accuracy</div><b>Accuracy</b> is the simplest score: out of the test machines, what fraction did the model get right? 0.90 means 9 out of 10 correct. Higher is better, but 100% on real data is usually a red flag, not a triumph.</div>

```python
predictions = model.predict(X_test)        # the model's guesses for the hidden machines

score = accuracy_score(y_test, predictions)
print('Accuracy on machines it had never seen:', round(score, 3))
print('In plain words: it got about', round(score*100), 'out of every 100 right.')
```

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-left:5px solid #2e7d32;background:#eaf5ea;padding:14px 18px;border-radius:6px;margin:10px 0;color:#222;line-height:1.55;font-size:15px;"><div style="font-weight:700;letter-spacing:.4px;color:#2e7d32;font-size:13px;margin-bottom:6px;text-transform:uppercase;">What just happened</div>Around <b>0.87</b> &mdash; roughly 87 out of 100 correct, on machines the model never trained on. Not perfect, and that is healthy: real equipment has surprises, and a model claiming 100% is usually cheating somehow. This is an honest, useful result.</div>

<div style="font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,Helvetica,Arial,sans-serif;border-bottom:3px solid #1f2d3d;padding-bottom:8px;margin:22px 0 6px 0;font-size:22px;font-weight:800;color:#1f2d3d;"><span style="background:#0b6e7a;color:#fff;border-radius:20px;padding:2px 12px;font-size:14px;margin-right:10px;">Step 9</span>Use it on a brand-new machine</div>

This is the whole point. A new machine arrives. We have its three numbers but no idea whether it will need service. We hand the numbers to the trained model and get an instant call.

```python
# A clearly worrying machine: old, hot, shaky.
new_machine = pd.DataFrame([{'hours_used': 5200, 'temperature': 86, 'vibration': 4.6}])

guess = model.predict(new_machine)[0]
confidence = model.predict_proba(new_machine)[0][1]   # model's probability of 'needs service'

print('New machine:', new_machine.to_dict('records')[0])
print('Prediction :', 'NEEDS SERVICE' if guess == 1 else 'looks fine')
print('Confidence in needs-service:', round(confidence*100), '%')
```

```python
# A clearly healthy machine: young, cool, smooth.
healthy = pd.DataFrame([{'hours_used': 900, 'temperature': 65, 'vibration': 2.1}])

guess = model.predict(healthy)[0]
confidence = model.predict_proba(healthy)[0][1]

print('New machine:', healthy.to_dict('records')[0])
print('Prediction :', 'NEEDS SERVICE' if guess == 1 else 'looks fine')
print('Confidence in needs-service:', round(confidence*100), '%')
```
